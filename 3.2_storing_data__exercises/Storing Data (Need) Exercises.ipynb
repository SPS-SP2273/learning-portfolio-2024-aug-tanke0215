### Exercise 1 (Total recall?)
1. Two similarities between lists and arrays.
- Both are used to store data.
- Two lists and two arrays can be combined in the same way.

2. Two differences between lists and arrays.
- When a list is multiplied by n, the list repeats itself n times. When an array is multiplied by n, each individual element in the array is multiplied by n.
- Multiple data types can exist within a list. Only one data type exists within one array.

3. What is a dictionary?
*A dictionary is a collection of data stored key-value pairs.*

### Exercise 3 (Index again)

```python
elements = ['Hydrogen',
            'Helium', 'Lithium',
            'Beryllium', 'Boron', 'Carbon',
            'Nitrogen', 'Oxygen',
            'Fluorine',
            'Neon']

print(elements[4]) # print the element at index 4 using forward indexing
print(elements[-6]) # print the element at index 4 from the end of the list using reverse indexing
```

### Exercise 4 (How many ones)

```python
numbers=[45, 60, 1, 30, 96, 1, 96, 57, 16, 1,
        99, 62, 86, 43, 42, 60, 59, 1, 1, 35,
        83, 47, 34, 28, 68, 23, 22, 92, 1, 79,
        1, 29, 94, 72, 46, 47, 1, 74, 32, 20,
        8, 37, 35, 1, 89, 29, 86, 19, 43, 61]

n = 0
counter = 0
while n != len(numbers):
    print(numbers[n])
    if numbers[n] == 1:
        counter += 1
    n += 1
    print(f"currently {counter}")              
```

### Exercise 5 (A Matter of Statistics)

```python
import numpy as np

x_1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
x_2 = [1, 8, 27, 64, 125, 216, 343, 512, 729, 1000]

np_x_1 = np.array(x_1)
np_x_2 = np.array(x_2)

print(np_x_1.mean())
print(np_x_2.mean())
print(np_x_1.std())
print(np_x_2.std())

# Finding Pearson's Correlation Coefficient
def CorrCoeff(x,y):
    global r    # Globalise r, taking it out of the function
    r = (len(x)*sum(x*y) - sum(x)*sum(y))/np.sqrt((len(x)*sum(x**2) - sum(x)**2)*(len(y)*sum(y**2) - sum(y)**2))

CorrCoeff(np_x_1, np_x_2)
print(r)

```

## Exercise 1 (Do you know why?)
```python
def greeting(name):
    if name == 'Batman':
        return 'Hello Batman! So, nice to meet you!'
    return f'Hello {name}!' 
```

The code above works as expected as when Python sees and executes the return keyword in a function, it jumps out of the function with the return value and any code after the return line is not read and executed. 

### Notes on ``return``
``return`` gives a value from within the function for use outside of the function, for examplem, assigning a value to another variable. 

````python
print(greeting('Batman'))
print(greeting('Superman'))
````

## Exercise 2 (Chubby or not)

````python
def calculate_bmi(weight,height):
    BMI = weight / (height ** 2)
    if BMI < 18.5:
        return print("Underweight")
    elif 18.5 <= BMI < 25:
        return f'Normal weight'
    elif 25 <= BMI < 30:
        return f'Overweight'
    else:
        return f'Obese'
    
calculate_bmi(70,1.8)
````

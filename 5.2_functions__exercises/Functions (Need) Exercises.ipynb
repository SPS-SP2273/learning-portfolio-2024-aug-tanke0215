## Exercise 1 (Do you know why?)

```python
def greeting(name):
    if name == 'Batman':
        return 'Hello Batman! So, nice to meet you!'
    return f'Hello {name}!'
```

The following code works as expected despite not having an else statement as Python interprets code line-by-line, so if name is not equal to 'Batman', then it will skip the if-statement and run the rest as normal. 

```python
print(greeting('Batman'))
print(greeting('Superman'))
```

## Exercise 2 (Chubby or not)

```python
def calculate_bmi(weight,height):
    BMI = weight / (height ** 2)
    if BMI < 18.5:
        return print("Underweight")
    elif 18.5 <= BMI < 25:
        return f'Normal weight'
    elif 25 <= BMI < 30:
        return f'Overweight'
    else:
        return f'Obese'
    
calculate_bmi(70,1.8)
```

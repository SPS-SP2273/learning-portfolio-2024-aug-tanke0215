## Exercise 1 (Celsius to Fahrenheit or Kelvin)

```python
def convert_celsius(temperature_celsius,target_scale = "Fahrenheit"): # Target scale is set to Fahrenheit by default.
    if target_scale == "Kelvin":
        temperature_kelvin = temperature_celsius + 273.15
        return f'{temperature_celsius} degrees Celsius is {temperature_kelvin:.2f} Kelvin'
    if target_scale == "Fahrenheit":
        temperature_fahrenheit = 9/5 * temperature_celsius + 32
        return f'{temperature_celsius} degrees Celsius is {temperature_fahrenheit:.2f} degrees Fahrenheit'
    else:
        return f'{temperature_celsius} degrees Celsius is {temperature_celsius} degrees Celsius'
    
print(convert_celsius(100,"Kelvin"))
print(convert_celsius(100)) 
```

## Exercise 2 (Fahrenheit to Celsius or Kelvin)

```python
def convert_fahrenheit(temperature_fahrenheit,target_scale = "Celsius"): # Target scale is set to Celsius by default.
    if target_scale == "Kelvin":
        temperature_kelvin = (temperature_fahrenheit - 32)*5/9 + 273.15
        return f'{temperature_fahrenheit} degrees Fahrenheit is {temperature_kelvin:.2f} Kelvin'
    if target_scale == "Celsius":
        temperature_celsius = (temperature_fahrenheit - 32)*5/9
        return f'{temperature_fahrenheit} degrees Fahrenheit is {temperature_celsius:.2f} degrees Celsius'
    else:
        return f'{temperature_fahrenheit} degrees Fahrenheit is {temperature_fahrenheit} degrees Fahrenheit'
print(convert_fahrenheit(212,"Kelvin"))
print(convert_fahrenheit(212))
```

## Exercise 3 (General Temperature Conversion) 

```python
def convert_temperature(temperature, source_scale, target_scale):
    if source_scale == "Celsius":
        return convert_celsius(temperature,target_scale)
    if source_scale == "Fahrenheit":
        return convert_fahrenheit(temperature,target_scale)

print(convert_temperature(212, "Fahrenheit", "Kelvin"))
```

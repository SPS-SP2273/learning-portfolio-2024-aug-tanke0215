### Exercise 3 (Timing Stuff)  

```python
squares = [i**2 for i in range(10)]
print(squares)
```

```python
squares = list()
for i in range(10):
    squares.append(i**2)
print(squares)
```

```python
%timeit squares = [i**2 for i in range(10)]
print(squares)
```

```python
%timeit squares = list()
for i in range(10):
    squares.append(i**2)
print(squares)
```

The %timeit output represents the time taken to execute a block of code. The time taken to execute the second code block is approximately 10 times faster than the first code block.

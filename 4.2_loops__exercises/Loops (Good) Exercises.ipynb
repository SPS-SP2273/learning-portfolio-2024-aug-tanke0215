# Loops (Good) Exercises
## Exercise 1 (Make me an odd list)

```python
for n in range(10):
    if n % 2 == 0:
        continue
    square = n**2
    print(square)
```

## Exercise 2 (Make me another odd list)

```python
[number**2 for number in range(10) if number % 2 != 0]
```

## Exercise 3 (Time me!)

```python
%%timeit -n 5000
squarenumbers = []
for n in range(10):
    if n % 2 == 0:
        continue
    squarenumbers.append(n**2)
```

```python
%timeit -n 5000 [number**2 for number in range(10) if number % 2 != 0]
```

The code using list comprehension is faster.

## Exercise 4 (A problem of decay)

```python
initial_quantity = original_quantity = 100
counter = 0

while True:
    initial_quantity *= 0.95
    counter += 1
    if initial_quantity < 0.5*original_quantity:
        print(f'It will take the sample {counter} years to reduce to half its original amount.')
        break
```

## Exercise 5 (Changes in CO<sub>2</sub>)

```python
import numpy as np
co2_data = [
    (2000, 369.55), (2001, 371.14), (2002, 373.28), 
    (2003, 375.80), (2004, 377.52), (2005, 379.80), 
    (2006, 381.90), (2007, 383.79), (2008, 385.60), 
    (2009, 387.43), (2010, 389.90), (2011, 391.65), 
    (2012, 393.85), (2013, 396.52), (2014, 398.65),
    (2015, 400.83), (2016, 404.24), (2017, 406.55), 
    (2018, 408.52), (2019, 411.44), (2020, 414.24)
]

for i in range(len(co2_data)):
    if co2_data[i][1] >= (co2_data[i - 1][1] + 3):
        print(f'{co2_data[i][0]} showed an increase of CO2 of 3 ppm or more compared to the previous year. The corresponding change in concentration is {co2_data[i][1] - co2_data[i-1][1]:.2f} ppm from the previous year.' )
```

```python
morethan3ppm = [i for i in range(len(co2_data)) if co2_data[i][1] >= (co2_data[i-1][1] + 3)]
for j in range(len(morethan3ppm)):
    print(f'{co2_data[morethan3ppm[j]][0]} showed an increase of CO2 of 3 ppm or more compared to the previous year. The corresponding change in concentration is {co2_data[morethan3ppm[j]][1] - co2_data[morethan3ppm[j]-1][1]:.2f} ppm from the previous year.' )
```

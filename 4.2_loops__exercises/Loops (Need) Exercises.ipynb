## Exercise 1 (Celsius to Fahrenheit) 


````python
temperatures_celsius = [
    0, 5, 10, 15, 20, 25,
    30, 35, 40, 45, 50
]
temperatures_fahrenheit = []

for i in range(len(temperatures_celsius)):
    temperatures_fahrenheit.append(9/5 * temperatures_celsius[i] + 32)

for j in range(len(temperatures_celsius)):
    print(f'{temperatures_celsius[j]}, {temperatures_fahrenheit[j]}')
````

## Exercise 2 (Multiplication table) 

````python
for letter in ['A', 'B', 'C']:
    for number in [1, 2, 3]:
        print(f'{letter}{number}', end='\t')
    print('\n')
````

```python 
# Expected output
1 : 1   2   3   4   5   
2 : 2   4   6   8   10  
3 : 3   6   9   12  15  
4 : 4   8   12  16  20  
5 : 5   10  15  20  25  
```

````python
for i in range(1,6):
    print(f'{i} : {i}', end='')
    for j in range(2,6):
        print(f'{i*j:3}', end='')
    print('\n')
````

## Exercise 3 (Simulating a Coin Flip Experiment)  

````python
import numpy as np
heads = 0 
probability = np.random.rand(10)  


for i in range(len(probability)):
    for j in range(len(probability)):
        if probability[j] > 0.5:
            heads = heads + 1
    print(f'Experiment {i}: No. of Heads = {heads}')
    heads = 0 
    probability = np.random.rand(10)
````
